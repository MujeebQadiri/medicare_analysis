### Exploratory Data Analysis

### Load in data

```python
import requests
import pandas as pd
import pathlib

if pathlib.Path("data/medicare_part_d_spending.csv").exists():
    df = pd.read_csv("data/medicare_part_d_spending.csv")
else:
    api_endpoint = "https://data.cms.gov/data-api/v1/dataset/7e0b4365-fd63-4a29-8f5e-e0ac9f66a81b/data"
    meta = requests.get("https://data.cms.gov/data-api/v1/dataset/7e0b4365-fd63-4a29-8f5e-e0ac9f66a81b/data?size=1").json()

    all_data = []
    offset = 0
    size = 1000

    while True:
        params = {"size": size, "offset": offset}
        response = requests.get(api_endpoint, params=params)
        batch = response.json()
        
        if not batch:
            break
        
        all_data.extend(batch)
        offset += size
        print(f"Fetched {len(all_data)} records...")
        df = pd.DataFrame(all_data)
        df.to_csv("data/medicare_part_d_spending.csv")
```

### Column Features

```python
from IPython.display import display
import pdfplumber
import pathlib

with pdfplumber.open(pathlib.Path("Medicare Part D Spending by Drug/2023/Medicare Part D Spending by Drug Data Dictionary 20250425_508.pdf")) as pdf:
    tables = []
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            tables.extend(table)

dictionary_df = pd.DataFrame(tables[2:], columns=tables[1])
dictionary_df = dictionary_df.dropna()
dictionary_df = dictionary_df[dictionary_df['Variable Name'] != 'Variable Name']

styled = (dictionary_df.style
    .set_properties(**{'text-align': 'left', 'white-space': 'pre-wrap'})
    .set_table_styles([
        {'selector': 'th', 'props': [('background-color', '#2c5f8a'), 
                                      ('color', 'white'), 
                                      ('text-align', 'left'),
                                      ('padding', '8px')]},
        {'selector': 'td', 'props': [('padding', '6px 12px'), 
                                      ('border-bottom', '1px solid #ddd')]},
        {'selector': 'tr:hover', 'props': [('background-color', '#f0f4f8')]}
    ])
    .hide(axis='index')
)

display(styled)
```
